```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('nepal_covid19_data.csv')
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='date', y='Confirmed', color='royalblue')
plt.title('Total Confirmed COVID-19 Cases in Nepal Over Time', fontsize=16)
plt.xlabel('Date')
plt.ylabel('Confirmed Cases')
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(True)
plt.show()
```

```python
df['daily_confirmed'] = df['Confirmed'].diff().fillna(0)
df['daily_deaths'] = df['Deaths'].diff().fillna(0)

sns.lineplot(data=df, x='date', y='daily_confirmed', label='Daily New Cases')
sns.lineplot(data=df, x='date', y='daily_deaths', label='Daily New Deaths')
plt.title('Daily New COVID-19 Cases & Deaths in Nepal')
plt.xticks(rotation=45)
plt.show()
```

```python
df['Active_calc'] = df['Confirmed'] - df['Deaths'] - df['Recovered']
sns.lineplot(data=df, x='date', y='Active_calc')
plt.title('Active COVID-19 Cases in Nepal')
plt.xticks(rotation=45)
plt.show()
```
